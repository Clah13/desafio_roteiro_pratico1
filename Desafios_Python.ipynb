## 2.1 M√©dia Simples

  C√≥digo simples e direto ao ponto para c√°lculo da m√©dia com os elementos fornecidos pelo exerc√≠cio.

  Considerando que os n√∫meros s√£o est√°ticos, a soma sempre ser√° 150 e a m√©dia correta a ser alcan√ßada √© 30.
  
  `150 / 5 = 30`.

````python
numeros = [10,20,30,40,50] #Valores fixos j√° estabelecidos no c√≥digo

# Fun√ß√£o para calcular a m√©dia
def media(elementos):

  calculo = sum(elementos) / len(elementos) #Fiz o 2.2 antes do 2.1. Aqui est√° mais enxuto.

  return calculo
print("==== Programa Calcula M√©dia ====")
print("A m√©dia √©:", media(numeros))
````

### 2.2 - M√©dia din√¢mica

  Considerando que √© um desafio, constru√≠ uma vers√£o que suporta m√∫ltiplos tamanhos de lista, que √© escolhida pelo usu√°rio. Al√©m disso, os n√∫meros gerados s√£o aleat√≥rios at√© 100.

````python
from random import randrange

# Vari√°vel de lista vazia
lista = []

# Fun√ß√£o para adicionar n√∫meros aleat√≥rios √† lista
def adiciona(nome_lista, tamanho):

  for i in range(tamanho): # Itera para adicionar os elementos na lista

    nome_lista.append(randrange(0, 101)) #101 n√£o incluso no random

  return nome_lista

# Soma e tira a m√©dia dos n√∫meros da lista
def calcula(nome_lista):

  soma = sum(nome_lista)
  print("Soma:", soma) # Imprime a soma pra fins de valida√ß√£o e teste

  media = soma / len(nome_lista) # Divide pelo tamanho da lista no c√°lculo da m√©dia
  '''
  Aqui eu tamb√©m poderia colocar como media = soma / tamanho (vari√°vel tamanho),
   mas optei por n√£o adicionar para evitar que eu tivesse que incluir o tamanho
   como par√¢metro nesta fun√ß√£o.
  Ao inv√©s disso, optei por usar o tamanho da lista j√° que estava dispon√≠vel para
   ser calculado. Isso tamb√©m permite manter o c√°lculo correto da m√©dia caso a
   lista seja modificada novamente por alguma outra fun√ß√£o durante a execu√ß√£o do
   programa - seja com a adi√ß√£o ou remo√ß√£o de elementos.
  '''

  return media

# Executa as fun√ß√µes e imprime
print("==== Programa Calcula M√©dia Din√¢mica ====")

# O usu√°rio escolhe o tamanho da lista a ser gerada
print("Digite o tamanho: ")
tamanho = int(input())

print("Valores adicionados:", adiciona(lista, tamanho))
print("M√©dia:", calcula(lista))
````

## 3.1 - C√°lculo de lista est√°tica

  Seguindo a mesma l√≥gica do exerc√≠cio 2.1, segue como eu faria o exerc√≠cio de c√≥digo simples e pr√°tico, caso a lista seja com valores est√°ticos ou inclusa diretamente no c√≥digo, sem o input do usu√°rio.

````python
# Valores que cada pessoa gastou

Joao = [300, 500, 200, 800]
Pedro = [200, 400, 500, 700]

# Fun√ß√£o para calcular o gasto (eu fiz esse neg√≥cio em 2 minutos depois do sofrimento no item 3.2 hahaha)

def calculaGasto(lista1, lista2):
  valorl1 = sum(lista1) # soma valores da lista 1 (Joao)
  valorl2 = sum(lista2) # soma valores da lista 2 (Pedro)

  if valorl1 > valorl2: # compara valores e declara vencedor
    vencedor = "Jo√£o! Parab√©ns, seu gast√£o!"

  elif valorl2 > valorl1: # compara valores e declara vencedor
    vencedor = "Pedro! Parab√©ns, seu gast√£o!"

  else: # caso de empate
    vencedor = "Vish, empatou! Hora de melhorar esse planejamento financeiro, hein..."

  # resultado
  print(f"Jo√£o gastou R${valorl1} no m√™s e Pedro gastou R${valorl2} no m√™s. Nosso vencedor nos gastos √©...\n {vencedor}")

# Main (execu√ß√£o da fun√ß√£o)
calculaGasto(Joao, Pedro)
````

## 3.2 - Registro din√¢mico de valores

  Segundo a mesma l√≥gica do t√≥pico 2.2, tentei me desafiar ao m√°ximo para que o programa funcionasse majoritariamente com o input do usu√°rio e considerasse elementos aprendidos em aula, como de fun√ß√µes, gerenciamento das listas, loops, adi√ß√£o e remo√ß√£o dos valores.

  Outros elementos dispon√≠veis na linguagem tamb√©m acabaram sendo implementados durante a minha pesquisa para melhorar o c√≥digo, como o try except, filtros, lambda, dentre outros.

````python
#Listas que armazenar√£o as informa√ß√µes de clientes e valores

cadastro01 = []
cadastro02 = []

# Fun√ß√£o telaInicial - loop while que itera as op√ß√µes do menu. A √∫nica forma de sair √© pressionando 5
def telaInicial():
  while True:
    print("""
    ---------------------------------------

    Por favor, selecione a op√ß√£o desejada:

    1 - Cadastrar-se
    2 - Registrar consumo
    3 - Estat√≠sticas de valores devidos
    4 - Limpar registros
    5 - Sair
    ---------------------------------------
    """)
    escolha = input("> ")

    # "Switch case": Cada op√ß√£o leva para outra fun√ß√£o. Uma para o cadastro da pessoa,
    # outra para o cadastro do consumo, outra pra fazer a soma do que foi
    # consumido por cada pessoa e por fim uma para apagar um ou mais valores registrados.

    match escolha:
      case "1":
        cadastro()

      case "2":
        cadastroConsumo()

      case "3":
        calcConsumo()

      case "4":
        apagar()

      case "5":
        print("At√© a pr√≥xima!")
        return False #quebra o loop

      case _: #gen√©rico - "outros"
        print("\nOp√ß√£o inv√°lida. Tente novamente.")
  return True

# Fun√ß√£o cadastro, que √© respons√°vel por for√ßar o usu√°rio a cadastrar, obrigatoriamente, duas pessoas.
# No in√≠cio, coloquei para cadastrar uma pessoa por vez, mas com o objetivo de terminar isso logo eu modifiquei
# o c√≥digo para que as duas pessoas fossem cadastradas uma ap√≥s a outra.

# Caso uma pessoa seja apagada, o cadastro s√≥ ser√° feito do ID que ficou em branco. :)
# Obs: ainda possui recursividade, mas estou programando desde 06h da manh√£ e agora s√£o 21:13. N√£o tenho mais energia pra modificar esse c√≥digo por while e testar.

def cadastro():

  if cadastro01 == []: # se estiver vazio
    print("\n~~ CADASTRO 01 ~~")
    nome = input("Por favor, digite o seu nome: > ")
    cadastro01.append(nome) #adiciona √† lista
    # ponto de aten√ß√£o: n√£o h√° valida√ß√£o do nome, ent√£o aqui aceita qualquer coisa, inclusive n√∫meros. Apesar de existir um filtro por int e float, e aqui ser tecnicamente string, n√£o testei se afeta a soma.

    # mensagem de feedback
    print("\n______________\n")
    print("Cadastro realizado com sucesso! Seu ID √© 01.")
    print("______________\n\n")
    print("Prosseguindo para o pr√≥ximo cadastro...\n\n")
    cadastro()

  elif cadastro02 == []: # se estiver vazio
    print("~~ CADASTRO 02 ~~")
    nome = input("Por favor, digite o seu nome: > ")
    cadastro02.append(nome) # adiciona √† lista
    # ponto de aten√ß√£o: n√£o h√° valida√ß√£o do nome, ent√£o aqui aceita qualquer coisa, inclusive n√∫meros. Apesar de existir um filtro por int e float, e aqui ser tecnicamente string, n√£o testei se afeta a soma.

    # mensagem de feedback
    print("\n______________\n")
    print("Cadastro realizado com sucesso! Seu ID √© 02.")
    print("______________\n\n")
    print("Voltando para a tela inicial...\n")

  else: # se nenhuma lista estiver vazia
    # mensagem de feedback
    print("\n______________\n")
    print("Cadastro de 02 pessoas realizado com sucesso!")
    print("______________\n\n")
    print("Voltando para a tela inicial...\n")

# Fun√ß√£o "cad√™ o cadastro?" que puxa a orelha do usu√°rio e retorna uma mensagem de erro gen√©rica
# se n√£o houve um cadastro pr√©vio nas op√ß√µes 02 e 03.
# Poderia implementar o mesmo na op√ß√£o 04 (apagar) mas n√£o deu tempo.

def cadeCadastro():

  print("\n****\nATEN√á√ÉO: N√£o √© poss√≠vel registrar nem calcular o consumo de pessoas n√£o cadastradas! Realize o cadastro de DUAS pessoas primeiro e tente novamente.\n****\n")
  cadastro() # chama a fun√ß√£o cadastro para completar o cadastro do usu√°rio.

# Fun√ß√£o cadastroConsumo, que √© respons√°vel por registrar o que foi gasto pelo usu√°rio.
# Primeiro, valida se o cadastro foi completo para depois sinalizar as op√ß√µes.
# A valida√ß√£o dos usu√°rios evita que o √≠ndice 0 fique nulo ou vazio.
# Ap√≥s o usu√°rio confirmar qual o cadastro, ele registra o valor gasto.
# √â poss√≠vel melhorar ainda mais o loop para que o mesmo usu√°rio cadastre novos valores repetidamente at√© que ele pe√ßa para sair, mas no momento vai ficar assim.

def cadastroConsumo():
  if cadastro01 == [] or cadastro02 == []:
    cadeCadastro()
    return

  while True:
    cadastralista = input("""
    ........................\n
    Por favor, selecione o usu√°rio a ter o consumo cadastrado:
    1 - """ + cadastro01[0] + """
    2 - """ + cadastro02[0] + """
    3 - Voltar\n
    ........................\n
    """) # O primeiro item da lista sempre ser√° o nome cadastrado do usu√°rio

    if cadastralista == "1":
      valor = input("\nSelecione o valor da despesa a ser cadastrada: ")
      try: # Try catch / try except para registros de valores que n√£o s√£o n√∫meros
        validacao = float(valor)

        if validacao < 0: # Valida√ß√£o adicional. Valor num√©rico tamb√©m n√£o pode ser negativo
          print("\nO valor n√£o pode ser menor que zero.")
          continue

        cadastro01.append(float(valor)) # Adiciona √† lista 01
        print("\nCadastro realizado com sucesso!")

      except ValueError: # Em caso de erro
        print ("\n** Formato inv√°lido. Por favor, tente novamente e utilize apenas n√∫meros com ponto na casa decimal.**\n")

    elif cadastralista == "2":
      valor = input("\nSelecione o valor da despesa a ser cadastrada: ")
      try: # Try catch / try except para registros de valores que n√£o s√£o n√∫meros
        validacao = float(valor)

        if validacao < 0: # Valida√ß√£o adicional. Valor num√©rico tamb√©m n√£o pode ser negativo
          print("\nO valor n√£o pode ser menor que zero.")
          continue

        cadastro02.append(float(valor)) # Adiciona √† lista 02
        print("\nCadastro realizado com sucesso!")

      except ValueError: # Em caso de erro
        print ("\n** Formato inv√°lido. Por favor, tente novamente e utilize apenas n√∫meros com ponto na casa decimal.**\n")

    elif cadastralista == "3": # Retornar para o menu
      print ("\n** Voltando para a tela inicial... **\n")
      return

    else: #Outras possibilidades de preenchimento incorreto
      print("\n_____________\n")
      print("Op√ß√£o inv√°lida. Tente novamente ou digite 3 para voltar.")
      print("_____________")

# Fun√ß√£o calcConsumo
# Basicamente filtra o que n√£o for n√∫mero inteiro ou float da nossa lista e soma os valores filtrados.
# Cont√©m tratativa de erros para cobrir possibilidade de valores zerados ou n√£o preenchidos.
# Informa ao usu√°rio quem dos usu√°rios registrados gastou mais ou se ambos gastaram o mesmo valor.

def calcConsumo():
  if cadastro01 == [] or cadastro02 == []: # Verifica se existe usu√°rio em branco
    cadeCadastro()

  consumo1 = sum(list(filter(lambda x: isinstance(x, (int, float)) and not isinstance(x, bool), cadastro01))) # Filtra os valores da lista 01 para mostrar apenas int e float. Como boleano √© considerado um n√∫mero [0,1], o filtra tamb√©m.
  # Soma os valores ap√≥s a filtragem.

  if consumo1 == 0: # Se nenhum valor for cadastrado, retornar√° zero. Valida se √© zero ou n√£o.
    print(cadastro01[0],", voc√™ n√£o possui gastos registrados!")

  else:
    print("\nO consumo de "+ cadastro01[0] + " √©:", consumo1) #Informa o consumo

  consumo2 = sum(list(filter(lambda x: isinstance(x, (int, float)) and not isinstance(x, bool), cadastro02))) # Filtra os valores da lista 02 para mostrar apenas int e float. Como boleano √© considerado um n√∫mero [0,1], o filtra tamb√©m.
  # Soma os valores ap√≥s a filtragem.

  if consumo2 == 0: # Se nenhum valor for cadastrado, retornar√° zero. Valida se √© zero ou n√£o.
    print(cadastro02[0],", voc√™ n√£o possui gastos registrados!")

  else: # Informa o consumo
    print("\nO consumo de " + cadastro02[0] + " √©:", consumo2)

  # Compara√ß√£o de valores
  if consumo1 > consumo2:
    print("\nA pessoa que mais gastou √©: " + cadastro01[0])

  elif consumo2 > consumo1:
    print("\nA pessoa que mais gastou √© " + cadastro02[0])

  else:
    print("\nO consumo foi do mesmo valor.")

  # Mensagem informativa avisando que os valores n√£o s√£o armazenados em nenhum lugar fixo/banco de dados.
  print("\nOs dados ser√£o automaticamente apagados ap√≥s a finaliza√ß√£o do programa.")

# Fun√ß√£o apagar
# Funciona em loop while e apaga todo o registro da lista selecionada.
# H√° valida√ß√£o caso a lista esteja vazia.
# Obs: o c√≥digo poderia ser melhorado colocando a valida√ß√£o da lista antes do usu√°rio
# selecionar a op√ß√£o desejada, para que ele apague apenas de fato o que precisar.
# Neste momento, mesmo com a lista vazia, h√° a op√ß√£o de apagar.

def apagar():
  while True:

    apagalista = input("""
    Por favor, selecione o ID do usu√°rio a ser apagado:
    1 - ID 01
    2 - ID 02
    3 - Voltar\n
    """)
    print() # Op√ß√£o por ID ao inv√©s de nome pois pode existir uma lista j√° vazia.

    if apagalista == "1": # Valida√ß√£o se existem valores a serem apagados.
      if len(cadastro01) == 0:
        print("\n_____________\n")
        print("N√£o h√° informa√ß√£o a ser apagada.")
        print("_____________")

      else:
        cadastro01.clear() # Comando de apagar a lista
        print("\n_____________\n")
        print("Opera√ß√£o realizada com sucesso!")
        print("_____________\n")

    elif apagalista == "2":

      if len(cadastro02) == 0: # Valida√ß√£o se existem valores a serem apagados.
        print("\n_____________\n")
        print("N√£o h√° informa√ß√£o a ser apagada.")
        print("_____________\n")

      else:
        cadastro02.clear() # Comando de apagar a lista
        print("\n_____________\n")
        print("Opera√ß√£o realizada com sucesso!")
        print("_____________\n")

    elif apagalista == "3": # Voltando pra tela inicial
      print("\n** Voltando para a tela inicial... **\n")
      return #Sai desse loop e volta para o da tela inicial.

    else: #Tratativa de erro para op√ß√µes inv√°lidas.
      print("\n_____________\n")
      print("Op√ß√£o inv√°lida. Tente novamente ou digite 3 para voltar.")
      print("_____________\n")

# ____________________________________
# Execu√ß√£o do c√≥digo principal (main)


print("======== REGISTRO E COMPARA√á√ÉO DE GASTOS 'DUPLA DIN√ÇMICA' ==========")

print()

print("Ol√°, seja bem-vindo(a) ao nosso sistema de controle de gastos!")
print("Recomenda-se como primeiro passo realizar o cadastro de DUAS pessoas.")

telaInicial() #Chama o loop da tela inicial.
print("___") # Toque para finaliza√ß√£o.
````
